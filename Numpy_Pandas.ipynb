They are also incredibly important to Machine Learning as they allow one to explore the data (and formulate the questions based on it) and also CLEAN/PREPARE the data. It is not an exageration saying that 80-90% of the work done by a Data Scientist is cleaning data.

[pandas](https://pandas.pydata.org/) is a powerful library that allows one to work with data.

Being able to see the entirety of the data on the screen is a double edged sword. In one hand the human working on a given dataset can be there visually inspecting the dataset all the time. It is then more practical to use a different approach when dealing with large datasets.<br> 

In this sense there are a few tools we can use to inspect the data for example [head](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.head.html) and [tail](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.tail.html).

```python
# Displaying the bottom 5 rows of the dataframe.
pokemon_df.tail()
```
