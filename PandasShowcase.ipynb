Numbers are without context completely deprived of meaning. 

For example, you could ask yourself "is Bulbasaur attack high?" but it is impossible to answer this without knowing the distribution of values of a given column.

A useful tool to obtain this information is [describe](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html)

```python
pokemon_df.describe()
```

For non-numerical columns other tools are more appropriated to understand the value distribution, such as

```python
# This shows all the unique entries in the dataset for that column.
pokemon_df['Type 1'].unique()
```

```python
# This will count how much incidence of a given unique entry is seen on the dataframe.
pokemon_df['Type 1'].value_counts()
```

It is also possible to query specific aspects of each individual column. To select only a column it is similar to getting a specific item of a dictionary,

```python
# Getting only the items that are in the HP column.
pokemon_df['HP']
```

```python
pokemon_df.HP
```

```python
# A list of columns can also be selected.
pokemon_df[['HP','Attack','Speed']]
```

As visual beings we are it is often useful to deploy plots in our journey to understand the dataset.

```python
pokemon_df['Attack'].plot()
```

```python
pokemon_df['Attack'][0:10].plot()
```

As it can be seen a line plot doesn't add much to the understanding of the column Attack. A histogram is much more appropriate to explore a column like this.

```python
pokemon_df['Attack'].hist()
```

```python
# For the type column we can use a pie chart for example.
pokemon_df['Type 1'].value_counts().plot(kind='pie')
```

While these plots already provide some information they are fairly simplistic and can be easily improved on.

```python
# Importing seaborn.
import seaborn as sns

sns.set(context='talk')
```

```python
# Creating a figure with subplots.
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(12,6))

# Plotting on the first axes.
pokemon_df['HP'].plot.hist(ax=axes[0], bins=50, alpha=0.5, color='red')
# Adding an individual title to this axis.
axes[0].set_title('HP')

# Plotting on the second axes.
pokemon_df['Defense'].plot.hist(ax=axes[1], bins=50, alpha=.5, color='red')
pokemon_df['Attack'].plot.hist(ax=axes[1], bins=50, alpha=.5, color='blue')
axes[1].set_title('Defense (red) and attack (blue)')

# Plotting on the third axes.
pokemon_df['Speed'].plot.hist(ax=axes[2], bins=50, alpha=0.5, color='green')
axes[2].set_title('Speed')

axes[1].set_ylabel('')
axes[2].set_ylabel('')
```

```python
# Defining the figure and axes.
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

# Creating the pie charts on their respective axes.
pokemon_df['Type 1'].value_counts().plot(kind='pie', ax=axes[0])
pokemon_df['Type 2'].value_counts().plot(kind='pie', ax=axes[1])

# "Removing" the y axis labels.
axes[0].set_ylabel('')
axes[1].set_ylabel('')

# Adding an overall title.
fig.suptitle('Pie charts of type 1 (left) and type 2 (right)')

# Removing unecessary white space.
plt.tight_layout()
```
