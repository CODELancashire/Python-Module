```python
import re

def remove_non_numerical(input_string):
    # Use a regular expression to match non-numerical characters and replace them with an empty string
    result = re.sub(r'[^0-9]', '', input_string)
    return result

# Example usage:
input_string = "abc123xyz456"
output_string = remove_non_numerical(input_string)
print(output_string)
```

```python
# Define a dictionary to map classifications to numerical quantities
class_to_number = {'Average Working/Obedience Intelligence':1, 
                   'Excellent Working Dogs':3,
                   'Fair Working/Obedience Intelligence':0,
                   'Above Average Working Dogs':2, 
                   'Brightest Dogs':4}
```

```python
# Create a new column by mapping the 'Classifications' column using the defined dictionary
df['NumericalClassification'] = df['Classification'].map(class_to_number)
```
